```python
import numpy as np
```

```python
rand_array = np.random.rand(80,3).astype(np.float64);
```

```python
out_str = '{\n'
for idx, bone in enumerate(rand_array):
    out_str += '{'
    out_str += str(bone[0])
    out_str += ','
    out_str += str(bone[1])
    out_str += ','
    out_str += str(bone[2])
    out_str += '}'
    if idx!=79 :
        out_str += ','
    out_str += '//color of bone[{}]'.format(idx)
    out_str += '\n'
out_str += '}'
    
```
